```python
# Nayelly Freire
#Professor Droescher 
#Quantitative Finance Spring 2023 
#Portfolio Analysis Project Assignment
#Due Feb 28

import pandas as pd 
import yfinance as yf

tickers=['aapl', 'msft', 'tsla', 'amc', 'meta', 'nio','f'] #tickers are stock symbols
etf=['spy','iwm','dia'] #the three given stocks we will compare them to later 
data=yf.download (['aapl', 'msft', 'tsla', 'amc', 'meta', 'nio','f'],period='2y', interval='3mo')['Adj Close'] #the yahoo finance data for the prices of each stocks
weights=[0.20]*7 # All 7 stocks have the same weight of 20%

returns=data.pct_change().dropna()
volatility = returns.std()*(252**0.5) #252 trading days in a year
```

```python
#Column 4,5,6 – Beta against SPY, IWM and DIA trailing 1 year

beta_data = []
for ticker in tickers:
    stock=yf.Ticker(ticker)

    try:
        beta_spy = stock.history(period="1y").Close.pct_change().cov(yf.Ticker("SPY").history(period="1y").Close.pct_change()) 
    except KeyError:
        beta_spy = None
        
    try:
        beta_iwm = stock.history(period="1y").Close.pct_change().cov(yf.Ticker("IWM").history(period="1y").Close.pct_change())
    except KeyError:
        beta_iwm = None
        
    try:
        beta_dia = stock.history(period="1y").Close.pct_change().cov(yf.Ticker("DIA").history(period="1y").Close.pct_change()) 
    except KeyError:
        beta_dia = None 
        
    beta_data.append((ticker, beta_spy, beta_iwm, beta_dia))
```

```python
#Column 7,8 – Average Weekly and Maximum Weekly Drawdown (52-week Low minus 52-week High) / 52-week High

drawdown_data = []

for ticker in tickers:
    stock = yf.Ticker(ticker)
    
    try:
        low = stock.history(period="1y").Low.min()
        high = stock.history(period="1y").High.max()
        drawdown=(low-high)/high
        avg_drawdown = drawdown.min()
        max_drawdown = drawdown.mean()
    except KeyError:
        avg_drawdown = None
        max_drawdown= None
```

```python
#Column 9,10- Total Return (using trailing 10-years), Annualized Total Return (using trailing 10-years)
returns_data = []

for ticker in tickers:
    stock = yf.Ticker(ticker)
    
    try:
        prices = stock.history(period="10y")['Close']
        total_return = prices.iloc[-1] / prices.iloc[0] - 1
        annualized_return = (1 + total_return)**(1/10) - 1
    except KeyError:
        total_return = None
        annualized_return = None
        
```
